```python
import folium

# Create a map centered around Sylhet, Bangladesh
m = folium.Map(location=[24.9, 91.9], zoom_start=9)

# Add a marker
folium.Marker(
    location=[24.9, 91.9],
    popup="Sylhet City",
    tooltip="Click me!"
).add_to(m)

# Add a simple polygon (like a fake flood extent)
folium.Polygon(
    locations=[
        [24.92, 91.88],
        [24.95, 91.89],
        [24.96, 91.93],
        [24.93, 91.95],
        [24.90, 91.92]
    ],
    color="blue",
    fill=True,
    fill_opacity=0.4,
    popup="Example Flood Area"
).add_to(m)

# Show the map
m
```
